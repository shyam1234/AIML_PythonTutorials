# Euler 3
The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143?

```python
# Using a function to check primality
def isPrime(n):
    if n == 2 or n == 3:
        return True
    if n % 2 == 0:
        return False
    r = 3
    while r * r <= n:
        if n % r == 0:
            return False
        r += 2
    return True
  
```

```python
isPrime(2), isPrime(7), isPrime(781), isPrime(169)
```

```python
N = 600851475143
```

```python
def factorize(n):
    factor = 2
    factors = []
    while n > 2:
        while n % factor == 0:
            factors.append(factor)
            n //= factor
        factor += 1
    return factors
```

```python
factorize(128)
```

```python
factorize(192)
```

```python
factorize(17)
```

```python
factorize(N)
```

```python
max(factorize(N))
```
